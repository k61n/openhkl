```python
import pynsx as nsx
import math
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
expt = nsx.Experiment('test', 'D19')
diff = expt.getDiffractometer()
reader = nsx.HDF5DataReader("/home/jonathan/data/d19_data/thom/y-corrected-data/125574.h5", diff)
data = nsx.DataSet(reader, diff)
expt.addData(data)

image = data.frame(150)
crop = image[100:200, 50:150]
plt.imshow(crop)
plt.show()
print(crop.mean(), crop.std()**2)
print(crop.std()**2 / crop.mean())
```
